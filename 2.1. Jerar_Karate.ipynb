```python
import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import pytensor
import random
import time
import os
import matplotlib.pyplot as plt
import seaborn as sns
import MULTISBM1IMPROVED as model

np.random.seed(42)
random.seed(42)
pytensor.config.default_rng = np.random.default_rng(42)

print("Cargando datos...")
df = pd.read_csv("Karate.csv")
data_matrix = df.to_numpy(dtype=int)
data_matrix = np.array(data_matrix, dtype=np.float64)
print(f"Tipo de data_matrix: {data_matrix.dtype}")
print(f"Forma de data_matrix: {data_matrix.shape}")

def evaluate_models_by_supblocks(data_matrix, min_supblocks=2, max_supblocks=3):
    """
    Evalúa el modelo jerárquico con diferentes números de superbloques y calcula métricas de información.
    
    Args:
        data_matrix: Matriz de datos de entrada
        min_supblocks: Número mínimo de superbloques a probar
        max_supblocks: Número máximo de superbloques a probar
    
    Returns:
        DataFrame con las métricas para cada número de superbloques
    """
    num_blocks = 4
    num_nodes = 34
    mu_zeta = 0  
    sigma_zeta = 2  
    a_tau = 2 
    b_tau = 1  
    a_alpha = 1
    b_alpha = 1
    a_sigma = 2
    b_sigma = 1 
    a_beta = 1
    b_beta = 1
    
    results = pd.DataFrame(columns=["num_supblocks", "WAIC", "AIC", "BIC", "time"])
    
    os.makedirs("traces", exist_ok=True)
    
    print(f"Comenzando evaluación para {max_supblocks - min_supblocks + 1} modelos...")

    for num_supblocks in range(min_supblocks, max_supblocks + 1):
        print(f"\n{'='*50}")
        print(f"Evaluando modelo con {num_supblocks} superbloques")
        print(f"{'='*50}")
        
        start_time = time.time()
        
        try:
            print("Creando modelo...")
            HierarchicalSBM = model.create_model_hierarchical_improved(
                data_matrix, 
                num_nodes=num_nodes, 
                num_blocks=num_blocks,
                num_supblocks=num_supblocks, 
                a_alpha=a_alpha,
                b_alpha=b_alpha,
                a_beta=a_beta,     
                b_beta=b_beta,    
                mu_zeta=mu_zeta,
                sigma_zeta=sigma_zeta,
                a_tau=a_tau,
                b_tau=b_tau,
                a_sigma=a_sigma,  
                b_sigma=b_sigma    
            )

            print(f"Iniciando muestreo MCMC para modelo con {num_supblocks} superbloques...")
            with HierarchicalSBM:
                init = 'adapt_diag'
                trace = pm.sample(
                    draws=10000,
                    tune=10000,
                    target_accept=0.95,
                    init=init,
                    return_inferencedata=True,
                    random_seed=42,
                    chains=4,
                    cores=2,
                    idata_kwargs={"log_likelihood": True},
                    progressbar=True  
                )
            
            print(f"Muestreo completado para {num_supblocks} superbloques")
   
            print("Calculando WAIC...")
            waic_data = az.waic(trace)
          
            print("Calculando AIC y BIC...")
            log_likelihood = trace.log_likelihood.Y_obs.values
            log_likelihood_sum = log_likelihood.sum()
          
            n_chains = trace.posterior.sizes["chain"]
            n_draws = trace.posterior.sizes["draw"]
   
            try:
                k = trace.posterior.dims["z"] + trace.posterior.dims["s"]  
            except KeyError:
                all_dims = trace.posterior.dims
                print(f"Dimensiones disponibles: {all_dims}")
                k = num_nodes + num_blocks * num_supblocks  
            
            n = data_matrix.size  
            
            # Calcular AIC y BIC
            aic = -2 * log_likelihood_sum + 2 * k
            bic = -2 * log_likelihood_sum + k * np.log(n)
            
            try:
                waic_value = -2 * float(waic_data.elpd_waic)
            except (AttributeError, KeyError):
                print("Estructura de waic_data:", waic_data)
                try:
                    waic_value = float(waic_data["waic"])
                except:
                    print("No se pudo extraer el valor WAIC, usando NaN")
                    waic_value = np.nan
            
            execution_time = time.time() - start_time
            
            results.loc[len(results)] = [num_supblocks, waic_value, aic, bic, execution_time]
 
            try:
                trace_path = f"traces/trace_supblocks_{num_supblocks}.nc"
                trace.to_netcdf(trace_path)
                print(f"Trace guardado en {trace_path}")
            except Exception as e:
                print(f"No se pudo guardar el trace para {num_supblocks} superbloques: {e}")
            
            print(f"\nTiempo de ejecución para {num_supblocks} superbloques: {execution_time:.2f} segundos")
            print(f"WAIC: {waic_value:.2f}, AIC: {aic:.2f}, BIC: {bic:.2f}")
 
            partial_results = results.copy()
            partial_results.to_csv(f"partial_results_upto_{num_supblocks}.csv", index=False)
            
        except Exception as e:
            print(f"Error al evaluar modelo con {num_supblocks} superbloques: {e}")
            import traceback
            traceback.print_exc()
            continue
    
    return results

def analyze_waic_results(results):
    """
    Analiza los resultados de la comparación de modelos usando WAIC y otras métricas.
    
    Args:
        results: DataFrame con los resultados de la evaluación
    """
    if len(results) == 0:
        print("No hay resultados para analizar.")
        return None
        
    os.makedirs("plots", exist_ok=True)
    
    print("Análisis de resultados de la comparación de modelos:")
    print("\nResumen de métricas por número de superbloques:")
    print(results)
    
    best_waic_idx = results["WAIC"].idxmin()
    best_aic_idx = results["AIC"].idxmin()
    best_bic_idx = results["BIC"].idxmin()
    
    best_waic = results.loc[best_waic_idx]
    best_aic = results.loc[best_aic_idx]
    best_bic = results.loc[best_bic_idx]
    
    print("\nMejor modelo según WAIC:")
    print(f"  Número de superbloques: {best_waic['num_supblocks']}")
    print(f"  WAIC: {best_waic['WAIC']:.2f}")
    
    print("\nMejor modelo según AIC:")
    print(f"  Número de superbloques: {best_aic['num_supblocks']}")
    print(f"  AIC: {best_aic['AIC']:.2f}")
    
    print("\nMejor modelo según BIC:")
    print(f"  Número de superbloques: {best_bic['num_supblocks']}")
    print(f"  BIC: {best_bic['BIC']:.2f}")
    
    sns.set_style("whitegrid")
    plt.figure(figsize=(12, 8))
    
    metrics = ["WAIC", "AIC", "BIC"]
    
    for i, metric in enumerate(metrics):
        plt.subplot(2, 2, i+1)
        sns.lineplot(x="num_supblocks", y=metric, data=results, marker='o', markersize=10)
        
        # Marcar el mejor valor
        best_idx = results[metric].idxmin()
        best_val = results.loc[best_idx]
        plt.scatter(best_val["num_supblocks"], best_val[metric], color='red', s=100, zorder=5)
        plt.annotate(f"Mejor: {best_val['num_supblocks']}", 
                     (best_val["num_supblocks"], best_val[metric]),
                     xytext=(10, -20), textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', color='red'))
        
        plt.title(f"{metric} por número de superbloques")
        plt.xlabel("Número de superbloques")
        plt.ylabel(metric)
        plt.xticks(results["num_supblocks"])
        plt.grid(True)
    
    plt.subplot(2, 2, 4)
    sns.barplot(x="num_supblocks", y="time", data=results)
    plt.title("Tiempo de ejecución por número de superbloques")
    plt.xlabel("Número de superbloques")
    plt.ylabel("Tiempo (segundos)")
    plt.xticks(range(len(results["num_supblocks"])), results["num_supblocks"])
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig("plots/supblocks_comparison.png", dpi=300)
    print("\nGráficos guardados en 'plots/supblocks_comparison.png'")
    
    best_model_n = int(best_waic["num_supblocks"])
    try:
        best_trace = az.from_netcdf(f"traces/trace_supblocks_{best_model_n}.nc")
   
        print("\nDiagnóstico de convergencia para el mejor modelo:")
        try:
            summary = az.summary(best_trace, var_names=["tau", "sigma"])
            print(summary[["mean", "sd", "r_hat"]])
        except:
            print("No se pudieron calcular estadísticas de convergencia para tau y sigma.")
   
        try:
            az.plot_trace(best_trace, var_names=["tau", "sigma"])
            plt.tight_layout()
            plt.savefig(f"plots/best_model_trace_supblocks_{best_model_n}.png", dpi=300)
            print(f"Diagnósticos guardados en 'plots/best_model_trace_supblocks_{best_model_n}.png'")
        except:
            print("No se pudieron generar gráficos de diagnóstico.")
        
    except FileNotFoundError:
        print(f"No se encontró el archivo de traza para el mejor modelo (superbloques={best_model_n}).")
    
    return best_waic["num_supblocks"]

min_supblocks = 2
max_supblocks = 3  

print(f"Evaluando modelos con {min_supblocks} a {max_supblocks} superbloques...")

results = evaluate_models_by_supblocks(
    data_matrix=data_matrix,
    min_supblocks=min_supblocks,
    max_supblocks=max_supblocks
)

results.to_csv("supblocks_comparison_results.csv", index=False)
print("\nResultados guardados en 'supblocks_comparison_results.csv'")
print("\nTabla de resultados:")
print(results)

best_supblocks = analyze_waic_results(results)
if best_supblocks is not None:
    print(f"\nSe recomienda usar {int(best_supblocks)} superbloques para el modelo final.")
```

```python
def calculate_manual_rhat(chains):
    """Calcula R-hat manualmente"""
    n_chains, n_samples = chains.shape
    
    chain_means = np.mean(chains, axis=1)
    overall_mean = np.mean(chain_means)
    
    B = n_samples * np.sum((chain_means - overall_mean)**2) / (n_chains - 1)
    
    chain_vars = np.var(chains, axis=1, ddof=1)
    W = np.mean(chain_vars)

    var_est = ((n_samples - 1) / n_samples) * W + B / n_samples
  
    r_hat = np.sqrt(var_est / W)
    return r_hat

def calculate_ess_simple(chains):
    """Estimación simple del ESS basada en autocorrelación lag-1"""
    n_chains, n_samples = chains.shape
    
    mean_autocorr = 0
    for chain in chains:
        norm_chain = (chain - np.mean(chain))
        autocorr = np.sum(norm_chain[1:] * norm_chain[:-1]) / np.sum(norm_chain**2)
        mean_autocorr += autocorr / n_chains
    
    ess = n_chains * n_samples * (1 - mean_autocorr) / (1 + mean_autocorr)
    return ess

def simple_diagnostics(log_posterior_values):
    """
    Calcula versiones simplificadas de ESS y R-hat
    """
    r_hat = calculate_manual_rhat(log_posterior_values)
    ess = calculate_ess_simple(log_posterior_values)
    
    print("\n===== DIAGNÓSTICOS DE CONVERGENCIA SIMPLIFICADOS =====")
    print(f"ESS estimado: {ess:.2f}")
    print(f"R-hat: {r_hat:.4f}")
    
    # Interpretación
    print("\n===== INTERPRETACIÓN =====")
    if ess < 400:
        print("⚠️ El ESS es bajo. Considere aumentar el número de iteraciones.")
    else:
        print("✓ El ESS es adecuado.")
        
    if r_hat > 1.01:
        print("⚠️ El R-hat es mayor a 1.01. Las cadenas no han convergido completamente.")
    else:
        print("✓ El valor de R-hat indica buena convergencia de las cadenas.")
    
    return ess, r_hat
```

##MATRIZ DE INCICENDIA

```python
start_idx = 0 
# Extraer las muestras de E_vector del trace
E_samples_full = trace_hier.posterior["E_vector"].values
print(f"Forma original de E_samples: {E_samples_full.shape}")
if len(E_samples_full.shape) == 3:
    E_samples_filtered = E_samples_full[:, start_idx:, :]
    num_chains = E_samples_filtered.shape[0]
    num_draws_per_chain = E_samples_filtered.shape[1]
    num_nodes = E_samples_filtered.shape[2]
    E_samples = E_samples_filtered.reshape(num_chains * num_draws_per_chain, num_nodes)
else:
    E_samples = E_samples_full.reshape(-1, E_samples_full.shape[-1])
    total_samples = E_samples.shape[0]
    samples_to_keep = total_samples // 2  
    E_samples = E_samples[total_samples - samples_to_keep:, :]

print(f"Forma final de E_samples para análisis: {E_samples.shape}")
print(f"Usando muestras desde la iteración: {start_idx}")

# ---- FUNCIONES AUXILIARES ----
def robust_mode(arr):
    """Encuentra el modo robusto de un array."""
    unique, counts = np.unique(arr, return_counts=True)
    max_count = np.max(counts)
    candidates = unique[counts == max_count]  
    return np.min(candidates)  

num_samples, num_nodes = E_samples.shape

# Create incidence matrix
incidence_matrix = np.zeros((num_nodes, num_nodes))
for sample in E_samples:
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):
            if sample[i] == sample[j]:  
                incidence_matrix[i, j] += 1
                incidence_matrix[j, i] += 1  

incidence_matrix /= num_samples
incidence_matrix_rounded = np.round(incidence_matrix, 2)

# Get mode assignments for each node
E_modes = np.array([robust_mode(E_samples[:, i]) for i in range(num_nodes)])

unique_blocks = np.unique(E_modes)
block_mapping = {block: i for i, block in enumerate(unique_blocks)}
E_modes_mapped = np.array([block_mapping[b] for b in E_modes])

# Sort by the mapped block assignments
sorted_indices = np.argsort(E_modes_mapped)
incidence_matrix_sorted = incidence_matrix_rounded[np.ix_(sorted_indices, sorted_indices)]
df_sorted = pd.DataFrame(incidence_matrix_sorted, 
                         index=sorted_indices+1, 
                         columns=sorted_indices+1)

block_summary = {}
for block in range(len(unique_blocks)):
    nodes_in_block = np.where(E_modes_mapped == block)[0] + 1  # +1 for 1-indexing
    block_summary[f"Block {block+1}"] = {
        "Nodes": nodes_in_block.tolist(),
        "Count": len(nodes_in_block)
    }

print("\nBloque Assignments Summary (muestras convergentes):")
for block, info in block_summary.items():
    print(f"{block}: {info['Count']} nodes - {info['Nodes']}")

# ---- VISUALIZACIÓN ----
plt.figure(figsize=(12, 10))
elegant_cmap = sns.color_palette("YlGnBu", as_cmap=True)
sns.heatmap(df_sorted, cmap=elegant_cmap, cbar=True, 
            linewidths=0.5, linecolor='white',
            cbar_kws={'label': 'Probabilidad de incidencia'})
plt.title(f"Matriz de Incidencia", 
          fontsize=16, fontweight='bold')
plt.xlabel("Nodos", fontsize=14)
plt.ylabel("Nodos", fontsize=14)
plt.tick_params(labelsize=10)

current_pos = 0
num_blocks_inferred = len(unique_blocks)
elegant_boundary_palette = sns.color_palette("Set2", num_blocks_inferred)

for block_idx in range(num_blocks_inferred):
    block_size = np.sum(E_modes_mapped[sorted_indices] == block_idx)
    if block_size > 0:
        plt.gca().add_patch(plt.Rectangle((current_pos, current_pos), 
                                      block_size, block_size, 
                                      fill=False, edgecolor=elegant_boundary_palette[block_idx], 
                                      lw=2.5, linestyle='-'))
        current_pos += block_size

plt.tight_layout()
plt.savefig(f"Incide_Jearqui_Karate.png", dpi=300, bbox_inches='tight')
plt.show()

df_sorted.to_csv(f"matriz_incidencia_desde_{start_idx}.csv")
np.save(f"asignaciones_bloque_desde.npy", E_modes_mapped)

df_assignments = pd.DataFrame({
    'Node': np.arange(1, num_nodes + 1),
    'Block': E_modes_mapped + 1  # +1 para que los bloques empiecen en 1
})
df_assignments.to_csv(f"asignaciones_nodo_bloque_desde_{start_idx}.csv", index=False)

print(f"Se han guardado los resultados en archivos para análisis posterior")
```

```python
gamma_samples = trace_hier.posterior["gamma_row"].values
total_samples_per_chain = gamma_samples.shape[1]
second_half_idx = total_samples_per_chain // 2
gamma_samples = gamma_samples[:, second_half_idx:, :]
gamma_samples = gamma_samples.reshape(-1, 4)  # (total_samples, num_blocks=4)
num_samples = gamma_samples.shape[0]
num_blocks = gamma_samples.shape[1]  # 4 bloques
num_superblocks = 2  

print(f"Shape of gamma_samples: {gamma_samples.shape}")
print(f"Usando {num_samples} muestras (segunda mitad de las cadenas)")

def get_high_certainty_samples(gamma_samples, certainty_threshold=0.8):
    """
    Identifica muestras donde las asignaciones tienen alta certeza.
    """
    sample_modes = []
    sample_certainties = []
    
    for sample_idx in range(gamma_samples.shape[0]):
        sample = gamma_samples[sample_idx]
        modes = {}
        for superblock in range(num_superblocks):
            count = np.sum(sample == superblock)
            modes[superblock] = count
        
        max_count = max(modes.values())
        certainty = max_count / num_blocks
        
        sample_modes.append(max(modes, key=modes.get))
        sample_certainties.append(certainty)
    
    sample_modes = np.array(sample_modes)
    sample_certainties = np.array(sample_certainties)
    
    high_certainty_indices = np.where(sample_certainties >= certainty_threshold)[0]
    
    print(f"Muestras de alta certeza: {len(high_certainty_indices)} de {len(sample_certainties)}")
    
    if len(high_certainty_indices) < 100:
        print("Advertencia: Pocas muestras de alta certeza, reduciendo umbral")
        if certainty_threshold > 0.6:
            return get_high_certainty_samples(gamma_samples, certainty_threshold - 0.1)
        return gamma_samples  
    
    return gamma_samples[high_certainty_indices]

filtered_gamma_samples = get_high_certainty_samples(gamma_samples, 0.75)
print(f"Usando {filtered_gamma_samples.shape[0]} muestras filtradas para análisis")

gamma_cooccurrence = np.zeros((num_blocks, num_blocks))
for sample in filtered_gamma_samples:
    for i in range(num_blocks):
        for j in range(num_blocks):
            if sample[i] == sample[j]:
                gamma_cooccurrence[i, j] += 1

gamma_cooccurrence = gamma_cooccurrence / filtered_gamma_samples.shape[0]
random_expectation = 1/num_superblocks
gamma_normalized = (gamma_cooccurrence - random_expectation) / (1 - random_expectation)
gamma_normalized = np.clip(gamma_normalized, 0, 1)  
gamma_enhanced = gamma_normalized
gamma_enhanced = np.round(gamma_enhanced, 2)
distance_matrix = 1 - gamma_cooccurrence
condensed_dist = squareform(distance_matrix)
Z = linkage(condensed_dist, method='ward')
cluster_labels = fcluster(Z, num_superblocks, criterion='maxclust') - 1
block_indices = np.arange(num_blocks) + 1 
cluster_summary = {}
for cluster in range(num_superblocks):
    blocks_in_cluster = block_indices[cluster_labels == cluster]
    cluster_summary[f"Superblock {cluster+1}"] = {
        "Blocks": blocks_in_cluster.tolist(),
        "Count": len(blocks_in_cluster)
    }

print("\nSuperblock Assignments (Natural Clustering):")
for cluster, info in cluster_summary.items():
    print(f"{cluster}: {info['Count']} blocks - {info['Blocks']}")

df_natural = pd.DataFrame(
    gamma_enhanced,
    index=block_indices,
    columns=block_indices
)
df_natural.to_csv("incidGammaSBM_Filtered_Normalized_Natural.csv", index=True, header=True)

fig = plt.figure(figsize=(16, 8))

ax1 = fig.add_subplot(1, 2, 1)
elegant_cmap = sns.color_palette("YlGnBu", as_cmap=True)

sns.heatmap(df_natural, cmap=elegant_cmap, cbar=True,
            linewidths=0.5, linecolor='white', ax=ax1,
            cbar_kws={'label': 'Probabilidad normalizada'})
ax1.set_title("Matriz de Incidencia asignaciones de bloques a superbloques", fontsize=14, fontweight='bold')
ax1.set_xlabel("Bloques", fontsize=12)
ax1.set_ylabel("Bloques", fontsize=12)
ax1.tick_params(labelsize=10)

elegant_boundary_palette = sns.color_palette("Set2", num_superblocks)
for cluster_id in range(num_superblocks):
    blocks_indices = np.where(cluster_labels == cluster_id)[0]
    
    if len(blocks_indices) > 0:
        blocks = blocks_indices + 1
        
        for i in blocks:
            for j in blocks:
                ax1.add_patch(plt.Rectangle((j-1, i-1), 1, 1, 
                                          fill=False, edgecolor=elegant_boundary_palette[cluster_id],
                                          lw=2, linestyle='-'))

ax2 = fig.add_subplot(1, 2, 2)
dendrogram(Z, labels=block_indices, leaf_rotation=90, ax=ax2)
ax2.set_title('Dendrograma de Agrupaciones Jerárquico', fontsize=14)
ax2.set_xlabel('Bloques', fontsize=12)
ax2.set_ylabel('Distancia', fontsize=12)
plt.tight_layout()
plt.savefig("incid_Gamma_Karate.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python


def calculate_ppp_values(stats: Dict[str, List[float]], real_stats: Dict[str, float]) -> Dict[str, float]:
    """
    Calcula los valores PPP (posterior predictive p-values) para las estadísticas dadas.
    
    Args:
        stats: Diccionario con listas de estadísticas calculadas de las redes simuladas
        real_stats: Diccionario con valores reales de referencia
        
    Returns:
        Diccionario con los valores PPP para cada estadística
    """
    ppp_values = {}
    
    for stat_name in stats.keys():
        # Filtrar valores NaN
        valid_stats = np.array([x for x in stats[stat_name] if not np.isnan(x)])
        
        # Calcular el PPP-value (proporción de simulaciones más extremas que el valor real)
        ppp_value = np.mean(valid_stats <= real_stats[stat_name])
        
        ppp_values[stat_name] = ppp_value
    
    return ppp_values

def plot_ppp_statistics(
    stats: Dict[str, List[float]], 
    real_stats: Dict[str, float],
    ppp_values: Dict[str, float],
    output_file: Optional[str] = "ppp_hierarchical_model.png"
) -> plt.Figure:
    """
    Visualiza las estadísticas con los valores PPP calculados.
    
    Args:
        stats: Diccionario con estadísticas calculadas
        real_stats: Diccionario con valores reales de referencia
        ppp_values: Diccionario con valores PPP calculados
        output_file: Nombre del archivo para guardar la figura
        
    Returns:
        Figura de matplotlib
    """
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    stat_names = ['densidad', 'transitividad', 'asortatividad']
    
    for i, stat_name in enumerate(stat_names):
        # Filtrar valores NaN
        valid_stats = np.array([x for x in stats[stat_name] if not np.isnan(x)])
        
        # Calcular estadísticas
        posterior_mean = np.mean(valid_stats)
        posterior_std = np.std(valid_stats)
        
        # Histograma con densidad de probabilidad
        axes[i].hist(valid_stats, bins=min(100, len(valid_stats)//50 + 10), 
                    density=True, alpha=0.7, color='darkgreen')
        
        # Valor real
        axes[i].axvline(real_stats[stat_name], color='crimson', linestyle='--', linewidth=2,
                       label=f'Real: {real_stats[stat_name]:.4f}')
        
        # Media posterior
        axes[i].axvline(posterior_mean, color='black', linestyle='-', linewidth=1.5,
                       label=f'Media: {posterior_mean:.4f} (±{posterior_std:.4f})')
        
        # PPP-value en el título
        axes[i].set_title(f'Distribución de {stat_name.capitalize()}\nvalor PPP: {ppp_values[stat_name]:.4f}', 
                         fontsize=12, fontweight='bold')
        axes[i].set_xlabel(stat_name.capitalize(), fontsize=10)
        axes[i].set_ylabel('Densidad', fontsize=10)
        axes[i].grid(alpha=0.3, linestyle='--')
        axes[i].legend(loc='best', frameon=True, framealpha=0.9)
    
    plt.suptitle(' ', fontsize=14, fontweight='bold', y=1.05)
    plt.tight_layout()
    
    # Guardar figura si se especificó un nombre de archivo
    if output_file:
        plt.savefig(output_file, dpi=300, bbox_inches='tight')
    
    return fig

def interpret_ppp_values(ppp_values: Dict[str, float]) -> Dict[str, str]:
    """
    Interpreta los valores PPP para cada estadística.
    
    Args:
        ppp_values: Diccionario con valores PPP calculados
        
    Returns:
        Diccionario con interpretaciones para cada estadística
    """
    interpretations = {}
    
    for stat_name, ppp in ppp_values.items():
        if 0.05 <= ppp <= 0.95:
            interpretations[stat_name] = "ACEPTABLE: El modelo reproduce adecuadamente esta estadística."
        elif ppp < 0.05:
            interpretations[stat_name] = "SUBESTIMACIÓN: El modelo genera valores consistentemente menores que el real."
        else:  # ppp > 0.95
            interpretations[stat_name] = "SOBRESTIMACIÓN: El modelo genera valores consistentemente mayores que el real."
    
    return interpretations

def main_hierarchical():
    # Parámetros reales
    real_stats = {
        'densidad': 0.13,
        'transitividad': 0.25,
        'asortatividad': -0.47
    }

    # Reconstrucción de matrices
    theta_samples = trace_hier.posterior['Theta_kl'].values
    interaction_matrices = reconstruct_interaction_matrix(theta_samples)

    # Generar redes
    print("Generando redes desde el modelo jerárquico...")
    networks = generate_networks(
        interaction_matrices, 
        num_nodes=34, 
        num_blocks=4, 
        networks_per_matrix=2000
    )

    # Cálculo de estadísticas
    print("Calculando estadísticas de red...")
    network_stats = compute_network_statistics(networks)

    # Calcular PPP-values
    print("Calculando PPP-values...")
    ppp_values = calculate_ppp_values(network_stats, real_stats)
    
    # Interpretar PPP-values
    interpretations = interpret_ppp_values(ppp_values)

    # Visualización con PPP-values
    print("Generando visualizaciones con PPP-values...")
    fig = plot_ppp_statistics(network_stats, real_stats, ppp_values, "ppp_hierarchical_model.png")
    plt.savefig("pppKARATEMULTI.png", dpi=300, bbox_inches="tight")
    plt.show()

    # Resumen de estadísticas con PPP-values
    summary = summarize_statistics(network_stats, real_stats)
    
    print("\n===== RESUMEN DE ESTADÍSTICAS CON PPP-VALUES (MODELO JERÁRQUICO) =====")
    for stat_name, values in summary.items():
        print(f"\n{stat_name.upper()}:")
        print(f"  Real:              {values['real']:.4f}")
        print(f"  Media Simulada:    {values['media']:.4f}")
        print(f"  Desviación Std:    {values['desv_std']:.4f}")
        print(f"  Error Absoluto:    {values['error_abs']:.4f}")
        print(f"  Error Relativo:    {values['error_rel']:.2f}%")
        print(f"  PPP-value:         {ppp_values[stat_name]:.4f}")
        print(f"  Interpretación:    {interpretations[stat_name]}")

if __name__ == "__main__":
    main_hierarchical()
```
